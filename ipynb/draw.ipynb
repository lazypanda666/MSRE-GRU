```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

models_left = ['OTF  ', 'ORE  ', 'Our  ']
models_right = ['OTF  ', 'ORE  ', 'Our  ']

ustc = {
    "Accuracy":  [0.7942, 0.4616, 0.9567],
    "Precision": [0.7952, 0.5034, 0.9574],
    "Recall":    [0.8102, 0.4752, 0.9587],
    "F1":        [0.7983, 0.4529, 0.9573],
}

unsw = {
    "Accuracy":  [0.8602, 0.4047, 0.9272],
    "Precision": [0.8480, 0.4354, 0.9132],
    "Recall":    [0.8235, 0.4602, 0.9298],
    "F1":        [0.8169, 0.3948, 0.9197],
}

colors = {
    "Accuracy":  "#08306B",
    "Precision": "#2171B5",
    "Recall":    "#4292C6",
    "F1":        "#6BAED6",
}

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("(A) USTC-TFC2016", "(B) UNSW-NB15"),
    horizontal_spacing=0.06
)

# ======================
# LEFT (USTC)
# ======================
for metric in ["Accuracy", "Precision", "Recall", "F1"]:
    fig.add_trace(go.Bar(
        x=ustc[metric],
        y=models_left,
        orientation='h',
        marker_color=colors[metric],
        name=metric,

        # ===== 数值放内部末尾 =====
        text=[f"{v:.4f}" for v in ustc[metric]],
        textposition='inside',
        insidetextanchor='end',

        textfont=dict(
            color='white',
            size=12
        ),

        cliponaxis=False
    ), row=1, col=1)

# ======================
# RIGHT (UNSW)
# ======================
for metric in ["Accuracy", "Precision", "Recall", "F1"]:
    fig.add_trace(go.Bar(
        x=unsw[metric],
        y=models_right,
        orientation='h',
        marker_color=colors[metric],
        name=metric,
        showlegend=False,

        text=[f"{v:.4f}" for v in unsw[metric]],
        textposition='inside',
        insidetextanchor='end',

        textfont=dict(
            color='white',
            size=12
        ),

        cliponaxis=False
    ), row=1, col=2)

# ======================
# Layout (tight SCI style)
# ======================
fig.update_layout(
    barmode='group',
    template='plotly_white',

    width=1050,
    height=420,

    # 更紧凑
    margin=dict(l=70, r=20, t=40, b=85),

    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.1,
        xanchor="center",
        x=0.5,
        font=dict(size=12)
    ),

    font=dict(
        family="Arial",
        size=13,
        color="black"
    )
)

# ======================
# Axes (SCI clean)
# ======================
fig.update_xaxes(
    showgrid=True,
    gridcolor="rgba(0,0,0,0.08)",
    zeroline=False,

    # 稍微压缩动态范围（增强对比）
    range=[0.3, 1.0]
)

fig.update_yaxes(
    showgrid=False,
    autorange="reversed"
)

fig.show()
```
